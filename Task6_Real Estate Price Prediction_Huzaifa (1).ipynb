### Dataset: Real_estate.csv

### Import the necessary libraries

```python
import piplite
await piplite.install('seaborn')
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from IPython.display import display
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

### Read the csv data into a pandas dataframe and display the first 5 samples

```python
data = pd.read_csv('Real estate.csv')
#display instead of print due to wide csv file
display(data.head())
```

### Show more information about the dataset

```python
display(data.info())
```

### Find how many samples are there and how many columns are there in the dataset

```python
shapes = data.shape
print(shapes)
```

### What are the features available in the dataset?

```python
features = data.keys()
#could have used a for loop to iterate the output cleaner/
display(features)
```

### Check if any features have missing data

```python
missing_data = data.isna().sum()
display(missing_data)
```

### Group all the features as Independent features in X

```python
X = data.iloc[:, :-1]
#Display>print to show it cleaner
display(X)
```

### Group feature(s) as dependent features in y

```python
y = data.iloc[:, -1]
display(y)
```

### Split the dataset into train and test data

```python
#train test split -> 95% training data & 5% testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05,  random_state=0)
```

### Choose the model (Linear Regression)

```python
from sklearn.linear_model import LinearRegression
```

### Create an Estimator object

```python
LRM = LinearRegression()
```

### Train the model

```python
LRM.fit(X_train, y_train)
```

### Apply the model

```python
y_pred = LRM.predict(X_test)
display(y_pred)
```

### Display the coefficients

```python
display(LRM.coef_)
```

### Find how well the trained model did with testing data

```python
r2 = r2_score(y_test, y_pred)
print("r2 Score : ", r2)
```
